```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
st = np.array([np.nan,0.871837,0.5714892,0.4212692,0.3003208,0.241519,0.1821863,0.1669971])
dy = np.array([np.nan,0.8808019,0.5695016,0.4484979,0.3060743,0.2354855,0.1819691,0.1679687])
gu = np.array([np.nan,0.8605468,0.5657601,0.4085463,0.3058068,0.2458916,0.1823791,0.1697563])
omp = np.array([24,12,8,6,4,3,2,1])
mpi = np.array([1,2,3,4,6,8,12,24])
```

```python
fig, axs = plt.subplots()

axs.plot(mpi, st, 'bo-')
axs.plot(mpi, dy, 'ro-')
axs.plot(mpi, gu, 'go-')
axs.set_ylim(0,1.0)
axs.set_ylabel('Seconds')
axs.grid(True)
axs.text(22.2, 0.9, 'a', fontsize=16) #, fontweight='bold') #, va='top')
```

```python
dydiff,gudiff=dy-st,gu-st
```

```python
plt.stem(mpi,dydiff,linefmt= 'b-',markerfmt='bo',use_line_collection=True)
plt.stem(mpi,gudiff,linefmt= 'r-',markerfmt='rd',use_line_collection=True)
plt.show()
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

```

```python
labels = ['1', '2', '3', '4', '6','8','12','24']
dydiff,gudiff=dy-st,gu-st

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(1,2,figsize=(12,4))
rects1 = ax[0].bar(x - width, st, width,color='r', label='Static')
rects2 = ax[0].bar(x, dy, width,color='b', label='Dynamic')
rects3 = ax[0].bar(x + width, gu, width,color='g', label='Guided')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax[0].set_ylabel('Seconds')
ax[0].set_xlabel('MPI Ranks')
ax[0].set_xticks(x)
ax[0].set_xticklabels(labels)
ax[0].legend()
ax[0].text(0.05, 0.95, 'a', fontsize=16,ha='center', va='center', transform=ax[0].transAxes)
ax[0].grid(True)

ax[1].stem(x,dydiff,linefmt= 'b-',markerfmt='bo',label='Dynamic',use_line_collection=True)
ax[1].stem(x,gudiff,linefmt= 'g-',markerfmt='gd',label='Guided',use_line_collection=True)

ax[1].set_ylabel('Seconds')
ax[1].set_xlabel('MPI Ranks')
ax[1].set_xticks(x)
ax[1].set_xticklabels(labels)
ax[1].legend()
ax[1].text(0.05, 0.95, 'b', fontsize=16,ha='center', va='center', transform=ax[1].transAxes)
ax[1].grid(True)
#autolabel(rects1)
#autolabel(rects2)
#autolabel(rects3)

fig.tight_layout()

plt.show()
#plt.savefig('./scheduling.png',dpi=300)
```
